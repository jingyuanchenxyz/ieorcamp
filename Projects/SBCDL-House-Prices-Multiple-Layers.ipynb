```python
from keras.datasets import boston_housing
(train_data, train_targets), (test_data, test_targets) = boston_housing.load_data()
```

# Normalization: of the data: Do not worry for now

```python
mean = train_data.mean(axis=0)
train_data -= mean
std = train_data.std(axis=0)
train_data /= std
test_data -= mean
test_data /= std
```

# Shape of the data

```python
train_data.shape
```

```python
train_targets.shape
```

# Functions to Build the Models

# Build a model with 1 hidden layers, 16 neurons, and relu activations.

```python
def build_model_1(shape):
  input = keras.Input(shape)
  l1 = layers.Dense(16, activation="relu")(input)
  output = layers.Dense(1)(l1)
  model = keras.Model(inputs=input,outputs=output)
  return model 
```

# Build a model with 2 hidden layers, 16 neurons, and relu activations.

```python
def build_model_2(shape):
    input = keras.Input(shape)
    l1 = layers.Dense(16, activation="relu")(input)
    l2 = layers.Dense(16,activation="relu")(l1)
    output = layers.Dense(1)(l2)
    model = keras.Model(inputs=input,outputs=output)
    return model
```

# Build a model with 3 hidden layers, 16 neurons, and relu activations.

```python
def build_model_3(shape):
  input = keras.Input(shape)
  l1 = layers.Dense(16, activation="relu")(input)
  l2 = layers.Dense(16,activation="relu")(l1)
  l3 = layers.Dense(16,activation="relu")(l2)
  output = layers.Dense(1)(l3)
  model = keras.Model(inputs=input,outputs=output)
  return model
```

# Function to Train the Model

```python
def train_model(model,inputs,outputs):
  model.compile(optimizer="rmsprop",loss="mse",metrics=["mae"])
  model.fit(inputs,outputs,epochs=100)
```

# Function to Predict

```python
def predict(model,data):
    predictions = model.predict(data)
    return predictions
```

```python
def show(prediction,label):
    print("the prediction is ",prediction)
    print("the correct value is ",label)
```

# Building a model

```python
model = build_model_3((train_data.shape[1],))
model.summary()
```

# Training the Model

```python
train_model(model,train_data,train_targets)
```

# Shape of the Testing Data

# Predict

```python
predictions = predict(model,test_data)
```

# Run the three models and compares the accuracy

```python
model_1 = build_model_1((train_data.shape[1],))
train_model(model_1, train_data, train_targets)

model_2 = build_model_2((train_data.shape[1],))
train_model(model_2, train_data, train_targets)

model_3 = build_model_3((train_data.shape[1],))
train_model(model_3, train_data, train_targets)
```

```python
results_1 = model_1.evaluate(test_data, test_targets)
results_2 = model_2.evaluate(test_data, test_targets)
results_3 = model_3.evaluate(test_data, test_targets)

print(f"Model 1 - Loss: {results_1[0]}, MAE: {results_1[1]}")
print(f"Model 2 - Loss: {results_2[0]}, MAE: {results_2[1]}")
print(f"Model 3 - Loss: {results_3[0]}, MAE: {results_3[1]}")
```
